```python
# Import important libarery 
import pandas as pd
import numpy as np
import random

# set random seed to 10 to have consistent results
random.seed(10)

# create a list of item names
mylist = []
for item in range(1,21):
    mylist.append("item " + str(item))
name_list = np.array(mylist)

# import dataset and add columm names to the imporated dataset
df = pd.read_csv('https://raw.githubusercontent.com/abdo-projects/data-science-portfolio/main/Dataset/Market_Basket_Optimisation.csv', sep= ',', names=name_list )

# Show the the tope 5 obersrvation of the dataset
df.head(5)
```

```python
# add user id attribute into the dataframe
df.insert(0, 'user_id', range(1000, 1000 + len(df)))

#show dataframe
df
```

```python
# Create a new dataframe by metling the orginal dataframe using user_id as index for the melt
df_new = df.melt(id_vars=["user_id"])

# Rename the columns of df_new: df2_tidy
df_new = df_new.rename(columns = {'variable': 'item_order', 'value': 'item'}, inplace = False)

#show the top 5 results
df_new
```

```python
# clean the dataset by dropping rows 
df_clean = df_new.dropna()

# count the NaN values in a column in pandas DataFrame
df_clean.isna().sum()
```

```python
# add a new column date purchased to the dataframe where numbers will be randomly generated by randint
def random_dates(start, end, n=5):
    start_u = start.value//10**9
    end_u = end.value//10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
start = pd.to_datetime('2015-01-01')
end = pd.to_datetime('2021-05-15')

# call the function to generate random date and time and assign it to dataframe
df_clean["date_purchased"] = random_dates(start, end, df_clean.shape[0])

#show the top 5 results
df_clean
```

```python
# import dataset and add columm names to the imporated dataset
df_review = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/amazon_reviews_big.csv', sep= ',')

#show the top 5 results
df_review
```

```python
# clean the dataset by dropping rows 
df_review = df_review.dropna()

#show the number of attributes and observationsdata.shape
df_review.shape
```

```python
# drop rows to match orginal dataframe
df_review = df_review.drop(df_review.index[29363:])

#show the number of attributes and observationsdata.shape
df_review.shape
```

```python
# shuffle review dataset
df_review = df_review.sample(frac=1).reset_index(drop=True)

# add user id attribute into the dataframe
df_review.insert(0, 'id', range(1000, 1000 + len(df_review)))
df_clean.insert(0, 'id', range(1000, 1000 + len(df_clean)))

# merage dataframes review and data
df_clean = pd.merge(df_clean, df_review, on='id')


#show the top 5 results
df_clean
```

```python
# count the NaN values in a column in pandas DataFrame
df_clean.isna().sum()
```

```python
# slice unique items from the orginal datafram and assign it to a new datafram list
itemlist = df_clean['item'].unique().tolist()
df_list = pd.DataFrame(data=itemlist, columns=["item"])

# add product id attribute into the dataframe
df_list.insert(0, 'product_id', range(5000, 5000 + len(df_list)))

# save datafram as csv to be exported to add item category manually 
df_list.to_csv("item_list.csv")

#show the top 5 results
df_list.head(5)
```

```python
# import dataset and add columm names to the imporated dataset
df_list_cleaned = pd.read_csv('https://abdo.tech/wp-content/uploads/2021/05/item_list_cleaned.csv', sep= ',')

#show the top 5 results
df_list_cleaned.head(5)
```

```python
# final dataframe after merging list dataframe and orginal dataframe
data = pd.merge(df_clean, df_list_cleaned, on='item')

# reorder and select columns related
data = data[['user_id', 'product_id', 'item', 'category_code', 'category','overall','summary','reviewText','item_order']]


#show the top 5 results
data.head(5)
```

```python
# save datafram as csv to be exported to add item category manually 
data.to_csv("data.csv")
```
